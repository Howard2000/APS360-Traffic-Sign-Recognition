**1.1 Introduction** 

Under the dynamic driving environments, one of the most important components of a self-driving vehicle system is being able to detect and safely respond to real-time traffic events, including correctly identifying traffic signs. To enhance the safety of both the passengers and pedestrians, our team is motivated to design and develop a deep learning model that can accurately and responsively detect and classify traffic signs. We believe deep learning is useful to improve the performance of the current traffic sign classifier systems. Specifically, by training the neural network of the model to classify shapes and colors of a wide variety of traffic signs using various datasets, we can attain a high accuracy and speedy traffic sign detector and classifier to contribute to the improvement of future automated driving systems and create safer driving environments.

**1.3 Background & Related Work**

With the increasing demand for safety in driving, many automobile companies came out with the solution to assist the drivers. The electric vehicle manufacturer Tesla developed Autopilot, a self-driving vehicle system, which is deep neural networks trained on problems ranging from perception to control https://www.tesla.com/en_CA/AI . The system detects stop signs, parking signs and so on, and responses to the detection. Another vehicle manufacturer Mercedes Benz developed Active Speed Limit Assist to recognize speed limit signs. The speed limit is not only displayed on the dashboard but also adapted in the cruise control https://www.mercedes-benz-mena.com/en/passengercars/mercedes-benz-cars/models/eqc/safety.pi.html/mercedes-benz-cars/models/eqc/safety/driving-assistance-gallery/speed-limit. Both systems are applications of applying neural networks and our neural network will detect all kinds of traffic signs.

**3.1 Ethical Considerations**

Due to the dynamic and simultaneous nature of the traffic environments, we have limitations on the existing datasets with a lack of challenging and unexpected conditions such as poor visibility of signs due to damages or severe weather conditions. Although these unpredictable situations are rare, inaccuracy of recognizing the traffic signs may lead to severe consequences of the passengers and pedestrians. Hence, to minimize the probability of causing accidents, we will design and use our traffic sign recognition system carefully and prioritize the safety of the passengers and pedestrians. Specifically, the system will not be completely automoated and will require the driver to make decisions when it is not so sure of its recognition result.
